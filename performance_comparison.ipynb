```python
import os
import seaborn as sns
import pandas as pd

from matplotlib import pyplot as plt
```

```python
list_path_all = os.listdir('results/all')
id = "all"
df_all = pd.DataFrame()
dict_all = {}
for path in list_path_all:
    df = pd.read_csv(os.path.join('results/all',path))
    # print(df)
    df.rename(columns={"Unnamed: 0": "id"}, inplace=True)
    df.id = id+"_"+path[:-4]
    dict_all[path[:-4]] = df
    df_all = pd.concat([df_all, df])
```

```python
list_path_idfc = os.listdir('results/base')
id = "idfc"
df_idfc = pd.DataFrame()
dict_idfc = {}
for path in list_path_idfc:
    df = pd.read_csv(os.path.join('results/base',path))
    # print(df)
    df.rename(columns={"Unnamed: 0": "id"}, inplace=True)
    df.id = id+"_"+path[:-4]
    dict_idfc[path[:-4]] = df
    df_idfc = pd.concat([df_idfc, df])
    
# 2 pour chaque fd00
```

```python
list_path_pca = os.listdir('results/pca')
id = "pca"
df_pca = pd.DataFrame()
dict_pca = {}
for path in list_path_pca:
    df = pd.read_csv(os.path.join('results/pca',path))
    # print(df)
    df.rename(columns={"Unnamed: 0": "id"}, inplace=True)
    df.id = id+"_"+path[:-4]
    dict_pca[path[:-4]] = df
    df_pca = pd.concat([df_pca, df])
    
# 3 pour chaque fd00
```

# FD001

```python
# Performance
performance_fd001 = pd.DataFrame()
# all feature
df_all = dict_all['fd001']

# PCA approach
list_df_pca = {k: v for k, v in dict_pca.items() if "fd001" in k}
df_pca_fd001 = pd.DataFrame()
for k, v in list_df_pca.items():
    df_pca_fd001 = pd.concat([df_pca_fd001, v])
    
performance_fd001 = pd.concat([df_all, df_pca_fd001])

# IDFC approach
list_df_idfc = {k: v for k, v in dict_idfc.items() if "fd001" in k}
df_idfc_fd001 = pd.DataFrame()
for k, v in list_df_idfc.items():
    df_idfc_fd001 = pd.concat([df_idfc_fd001, v])
    
performance_fd001 = pd.concat([performance_fd001, df_idfc_fd001])

# Créer une correspondance entre les anciennes et nouvelles étiquettes
renaming_dict = {
    "all_fd001": "all",
    "pca_fd001": "PCA_lambda",
    "pca_fd001_nbcomp": "PCA_K",
    "pca_fd001_varimax": "PCA_varimax",
    "idfc_fd001": " IDFC_K",
    "idfc_fd001_bruit": "IDFC_K+1"
}

# Remplacer les valeurs de la colonne "id" par les nouvelles étiquettes
performance_fd001["method"] = performance_fd001["id"].replace(renaming_dict)
```

```python
plt.figure(figsize=(14, 4))
plt.subplot(1,3,1)
sns.boxplot(data=performance_fd001, x="method", y="RMSE", hue="method")
plt.xticks(rotation=45)
plt.title("RMSE by case on subset FD001")
plt.tight_layout()
plt.xlabel("")
# plt.grid(True, axis='y')

plt.subplot(1,3,2)
sns.boxplot(data=performance_fd001, x="method", y="S_score", hue="method")
plt.xticks(rotation=45)
plt.title("S_score by case on subset FD001")
plt.xlabel("")
plt.tight_layout()
# plt.grid(True, axis='y')

plt.subplot(1,3,3)
sns.boxplot(data=performance_fd001, x="method", y="time", hue='method')
plt.xticks(rotation=45)
plt.title("Time by case on subset FD001")
plt.xlabel("")
plt.tight_layout()
# plt.grid(True, axis='y')
plt.savefig("images/performances/perf_fdd001.png")
plt.show()
```
